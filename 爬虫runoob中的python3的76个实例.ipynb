```python
import requests
from bs4 import BeautifulSoup

import pdb

import codecs
import os
from tqdm import tqdm

import pandas as pd 
  
```

```python
def getCatalog(root,url):
    """
    获取python3实例的各个链接地址
    
    参数
    ----
    root: str 根目录
    url: str 目录地址
    
    返回
    ----
    urls：list str 各个链接的地址
    urls_name: list str 各个链接的名字
    """
    headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:73.0) Gecko/20100101 Firefox/73.0'}
    req = requests.get(url,headers=headers)
    html = req.content.decode('utf-8')
    soup = BeautifulSoup(html)
    root_div = soup.find('div',id="content")
    all_a = root_div.findAll('a')
    urls = []
    urls_name =[]
    print('获取目录链接中。。。')
    for a in tqdm(all_a):
        old_url = a['href']
        if old_url.startswith('/'):
            old_url = root+old_url[9:]
        else:
            old_url = root+old_url
        urls.append(old_url)
        urls_name.append(a.text)
    return urls,urls_name

def getCode(df,all_urls):
    """
    获取python3实例的详细代码 并保存到df里面去
    
    参数
    ----
    df: DataFrame 数据框
    all_urls: list str 各个链接的地址
    
    无返回
    """
    headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:73.0) Gecko/20100101 Firefox/73.0'}
    print('爬取各个网站的代码中')
    for i in tqdm(range(len(all_urls))):
        try:
            url = all_urls[i]
            req = requests.get(url,headers=headers)              
            html = req.content.decode('utf-8')
            soup = BeautifulSoup(html)
            #pdb.set_trace()
            example_code = soup.find(attrs={'class':'example_code'}).text
            df.at[i,'代码'] = example_code
        except Exception as e:
            print(e)
            print(url+'解析出错')
    print('所有网站已爬取完成！！！')

def writeCode(df,names):
    """
    将df中保存的代码，写入本地当中
    
    参数
    ----
    df: DataFrame 数据框
    names: list str 各个文件的名字
    
    无返回
    """
    if os.path.exists('python3实例') == False:
        os.mkdir('python3实例')
    print('正在保存代码中。。。')
    for i in tqdm(range(len(names))):
        file = names[i]+'.py'
        if df.at[i,'代码'] == None:
            print(name[i]+'没有保存成功，请仔细检查一下')
            continue
        if file.find('/') != -1:
            file = file.replace('/','_')
        file = 'python3实例/'+file
        if not os.path.exists(file): 
            with codecs.open(file,'w',encoding='utf-8') as f:
                f.write(str(df.at[i,'代码']))
    print('完成保存代码，保存在“python3实例”中，请查收！！！')
def main():
    """
    程序运行入口
    """
    url = 'https://www.runoob.com/python3/python3-examples.html'
    root = 'https://www.runoob.com/python3/'
    all_urls,all_names = getCatalog(root,url)
    df = pd.DataFrame(data={'网址':all_urls,'文件名':all_names})
    getCode(df,all_urls)
    writeCode(df,all_names)
```

```python
main()
```
